```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("happiness_data_1000.csv")
df
```

```python
x=df.drop("Happiness_Score",axis=1)


y=df["Happiness_Score"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pre=model.predict(x_test)
y_pre
```

```python
mse=mean_squared_error(y_pre,y_test)
rmse=np.sqrt(mse)
print(mse)
print(rmse)
score=r2_score(y_pre,y_test)
print(f"accuracy score :{score*100:.2f}%")
```

```python
plt.figure(figsize=[10,6])
plt.scatter(y_pre,y_test,c="k")
plt.plot([y_test.min(),y_test.max()],[y_pre.min(),y_pre.max()])
plt.show()
```

```python
new=np.array([[30275.94,0.19,51.78,0.67,0.26,0.39]])
predict=model.predict(new)
predict
plt.figure(figsize=[10,6])
plt.scatter(y_pre,y_test,c="k")
plt.plot([y_test.min(),y_test.max()],[y_pre.min(),y_pre.max()])
plt.plot(new,predict)
plt.show()
```
