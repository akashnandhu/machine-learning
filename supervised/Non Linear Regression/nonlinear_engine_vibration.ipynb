```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
df=pd.read_csv("nonlinear_engine_vibration.csv")
df
```

```python
X=df["Engine RPM"].values
y=df["Vibration Amplitude (mm/s)"].values
```

```python
def exp_model(x,a,b,c):
    return a*np.exp(b*x)+c
```

```python
param,p=curve_fit(exp_model,X,y,p0=(3500,-0.015,200))
```

```python
y_pred=exp_model(X,*param)
y_pred
```

```python
plt.figure(figsize=[10,6])
plt.scatter(X,y)
plt.scatter(X,y_pred)
plt.show()
```
