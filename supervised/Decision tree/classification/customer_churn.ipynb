```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score
```

```python
# pip install scikit-learn
```

```python
df=pd.read_csv("customer_churn_dataset_1000.csv")
df
```

```python
label=LabelEncoder()
df["gender"]=label.fit_transform(df["gender"])
df["subscription_type"]=label.fit_transform(df["subscription_type"])
df["contract_type"]=label.fit_transform(df["contract_type"])
df["payment_method"]=label.fit_transform(df["payment_method"])
df["payment_method"]=label.fit_transform(df["payment_method"])
df["has_internet_service"]=label.fit_transform(df["has_internet_service"])
df


```

```python
x=df.drop("churn",axis=1)
y=df["churn"]
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
accuracy=accuracy_score(y_pred,y_test)
print(f"accuracy:{accuracy *100:.2f}%")
```

```python
plt.figure(figsize=(8,10))
plot_tree(model,feature_names=["age","gender","subscription_type","contract_type","tenure_months","avg_monthly_spend","customer_support_calls","has_internet_service","payment_method"],class_names=["play","no play"],filled=True)
plt.show()
```
