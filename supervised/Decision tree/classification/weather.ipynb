```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree  import DecisionTreeClassifier,plot_tree 
```

```python
df=pd.read_csv("weather.csv")
df
```

```python
label=LabelEncoder()
df["Date.Full"]=label.fit_transform(df["Date.Full"])

df["Station.City"]=label.fit_transform(df["Station.City"])
df["Station.Code"]=label.fit_transform(df["Station.Code"])
df["Station.Location"]=label.fit_transform(df["Station.Location"])
df["Station.State"]=label.fit_transform(df["Station.State"])

df
```

```python
x=df.drop("Data.Wind.Speed",axis=1)
y=df["Data.Wind.Speed"]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=43)
x_train
```

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
accuracy=accuracy_score(y_pred,y_test)
print(f"accuracy:{accuracy*100:.2f}%")
```
